# 8.Calculate IQR(difference between 75% and 25% quartile)

```python
# PRINT IQR OF MONTHLY HH INCOME
iqr_Mthly_HH_Income = df['Mthly_HH_Income'][df['Mthly_HH_Income'].between(df['Mthly_HH_Income'].quantile(.25), df['Mthly_HH_Income'].quantile(.75), inclusive=True)]
print(iqr_Mthly_HH_Income)
```
